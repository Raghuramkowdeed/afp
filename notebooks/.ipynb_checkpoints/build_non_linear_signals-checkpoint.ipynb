```python
pm = PortfolioOptimizer(fac_data_dir='/Users/raghuramkowdeed/Desktop/afp_data/1000/factor_exposure_data/',
                        fac_ret_file = '/Users/raghuramkowdeed/Desktop/afp_data/1000/fac_ret_file.csv',
                        fac_cov_hl = 12.0,
                        fac_names = ['beta']
                       )
```

```python
l_sig_df = pd.DataFrame()
nl_knn_sig_df = pd.DataFrame()
#nl_kernel_sig_df = pd.DataFrame()
sharpe_df = pd.DataFrame()

for sig_name in sig_names :
    if sig_name in ['beta', 'idiovol'] :
        continue
    print(sig_name)
    this_sig_sr = sig_df[sig_name]
    this_sig_sr = pm.neu_signal(this_sig_sr)
    this_sig_sr = this_sig_sr.fillna(this_sig_sr.mean())
    this_ret_sr = sig_df['ret']
    
    try :
        l_sig_sr = get_linear_sig(sig_sr=this_sig_sr, ret_sr=this_ret_sr, look_back=look_back)
        nl_knn_sig_sr = get_knn_non_linear_sig(sig_sr=this_sig_sr, ret_sr=this_ret_sr, look_back=look_back)
        #nl_kernel_sig_sr = get_kernel_non_linear_sig(sig_sr=this_sig_sr, ret_sr=this_ret_sr, look_back=look_back)
        
        pnl_sr_org = pm.run_signal(sig_sr=this_sig_sr,neu_sig=True)
        pnl_sr_l= pm.run_signal(sig_sr=l_sig_sr,neu_sig=True)
        pnl_sr_knn = pm.run_signal(sig_sr=nl_knn_sig_sr,neu_sig=True)
        #pnl_sr_kernel = pm.run_signal(sig_sr=nl_kernel_sig_sr,neu_sig=True)
        
    
        org_shrp = pnl_sr_org.mean()*sqrt(12)/pnl_sr_org.std()  
        l_shrp = pnl_sr_l.mean()*sqrt(12)/pnl_sr_l.std()
        knn_shrp = pnl_sr_knn.mean()*sqrt(12)/pnl_sr_knn.std()
        #kernel_shrp = pnl_sr_kernel.mean()*sqrt(12)/pnl_sr_kernel.std()
    
        #result = pd.Series( [org_shrp,l_shrp, knn_shrp, kernel_shrp], index=['org','lm', 'knn', 'kernel'], name = sig_name)
        result = pd.Series( [org_shrp,l_shrp, knn_shrp], index=['org','lm', 'knn'], name = sig_name)
        sharpe_df = sharpe_df.append(result)
        l_sig_df[sig_name] = l_sig_sr
        nl_knn_sig_df[sig_name] = nl_knn_sig_sr
        #nl_kernel_sig_df[sig_name] = nl_kernel_sig_sr
        print(result)
        
    except Exception as e:
        print(e)
        print(sig_name + ' failed')
        continue

l_sig_df.index.name = 'date'
nl_knn_sig_df.index.name = 'date'
#nl_kernel_sig_df.index.name = 'date'

l_sig_df.to_csv('/Users/raghuramkowdeed/Desktop/afp_data/1000/l_sig_df_'+str(look_back)+'.csv')
nl_knn_sig_df.to_csv('/Users/raghuramkowdeed/Desktop/afp_data/1000/nl_knn_sig_df_'+str(look_back) +'.csv')
#nl_kernel_sig_df.to_csv('/Users/raghuramkowdeed/Desktop/afp_data/1000/nl_kernel_sig_df_'+str(look_back) +'.csv')
```
